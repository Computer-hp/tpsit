1. Scrivi una funzione che, dato un array di numpy, verifichi se i valori dell'array sono in ordine crescente, decrescente oppure nessuno dei due

```python
import numpy as np

def verifica_ordine(array):
    crescente = np.all(array[:-1] <= array[1:])
    decrescente = np.all(array[:-1] >= array[1:])

    if crescente:
        return "crescente"
    elif decrescente:
        return "decrescente"
    else:
        return "nessuno dei due"


array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([5, 4, 3, 2, 1])
array3 = np.array([1, 3, 2, 4, 5])

print(verifica_ordine(array1))  # crescente
print(verifica_ordine(array2))  # ecrescente
print(verifica_ordine(array3))  # nessuno dei due
```

2. Disegna un bersaglio 10x10, le cornici concentriche hanno valori crescenti verso il centro

3. Trova il valore più frequente in un array di numpy, usa i metodi unique e argmax

```python
import numpy as np

def frequenza(array):
    valori_unici, conteggi = np.unique(array, return_counts=True)
    indice_piu_frequente = np.argmax(conteggi)
    return valori_unici[indice_piu_frequente]



array1 = np.array([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
array2 = np.array([5, 5, 6, 6, 6, 7, 7, 7, 7, 8, 8, 8, 8, 8])
array3 = np.array([1, 1, 1, 2, 2, 3])

print(frequenza(array1))
print(frequenza(array2))
print(frequenza(array3))
```

7. Scrivi un programma che trovi tutti i numeri primi in un intervallo dato dall'utente.

```python
def is_prime(n):
    """Verifica se un numero è primo."""
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    max_divisor = int(n**0.5) + 1
    for d in range(3, max_divisor, 2):
        if n % d == 0:
            return False
    return True

def find_primes_in_range(start, end):
    """Trova tutti i numeri primi nell'intervallo [start, end]."""
    numeri = []
    for num in range(start, end + 1):
        if is_prime(num):
            numeri.append(num)
    return numeri

def main():
    start = int(input("Inserisci il valore iniziale dell'intervallo: "))
    end = int(input("Inserisci il valore finale dell'intervallo: "))
    numeri = find_primes_in_range(start, end)
    print(f"Numeri primi nell'intervallo da {start} a {end}: {numeri}")

if __name__ == "__main__":
    main()
```

8. Genera i primi 14 numeri della serie di Fibonacci usando numpy e fanne il grafico con
matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_fibonacci(n):
    if n <= 0:
        return np.array([])
    elif n == 1:
        return np.array([0])
    elif n == 2:
        return np.array([0, 1])

    fibonacci_series = np.zeros(n, dtype=int)
    fibonacci_series[0] = 0
    fibonacci_series[1] = 1
    for i in range(2, n):
        fibonacci_series[i] = fibonacci_series[i - 1] + fibonacci_series[i - 2]
    return fibonacci_series

# Generare i primi 14 numeri della serie di Fibonacci
n = 14
fibonacci_series = generate_fibonacci(n)

# Stampare la serie di Fibonacci
print("I primi 14 numeri della serie di Fibonacci:", fibonacci_series)

# Tracciare il grafico della serie di Fibonacci
plt.plot(fibonacci_series, marker='o', linestyle='-', color='b', label='Fibonacci')
plt.grid(True)
plt.show()
```

14. Definire una funzione che riceva due liste qualsiasi, e determini se tutti gli elementi della
prima lista si trovino anche nella seconda, restituendo il valore True oppure False.

```python
def confronta_liste(lista1, lista2):

    for elemento in lista1:
        if elemento not in lista2:
            return False
    return True

# Esempi di utilizzo della funzione
lista1 = [1, 2, 3]
lista2 = [3, 2, 1, 4, 5]
print(confronta_liste(lista1, lista2))  # True

lista3 = ['a', 'b', 'c']
lista4 = ['c', 'b', 'a', 'd', 'e']
print(confronta_liste(lista3, lista4))  # True

lista5 = [1, 2, 3]
lista6 = [4, 5, 6]
print(confronta_liste(lista5, lista6))  # False
```

15. genera un file CSV chiamato weather_data.csv con due colonne: Data e Temperatura
genera i dati della temperatura rilevata in 100 giorni di monitoraggio
Carica i dati meteorologici da un file CSV utilizzando np.loadtxt().
Esplora i dati visualizzando le prime righe e determinando il numero di righe e colonne.
Calcola la temperatura media, massima e minima registrata.
Calcola la deviazione standard della temperatura.
Trova la data in cui è stata registrata la temperatura massima.
Trova la data in cui è stata registrata la temperatura minima.
Visualizza un grafico della temperatura nel tempo

```python
import numpy as np
import csv
import random
from datetime import datetime, timedelta

# Genera dati casuali per 100 giorni di monitoraggio
def genera_dati():
    data_inizio = datetime(2024, 1, 1)
    giorni = [(data_inizio + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(100)]
    temperature = [round(random.uniform(10, 30), 1) for _ in range(100)]
    return giorni, temperature

# Crea il file CSV con i dati generati
def crea_file_csv(nomi_file, giorni, temperature):
    with open(nomi_file, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Data', 'Temperatura'])
        for data, temperatura in zip(giorni, temperature):
            writer.writerow([data, temperatura])

# Genera i dati e crea il file CSV
giorni, temperature = genera_dati()
crea_file_csv('weather_data.csv', giorni, temperature)

# Carica i dati dal file CSV
data = np.loadtxt('weather_data.csv', delimiter=',', skiprows=1, dtype={'names': ('Data', 'Temperatura'),
                                                                       'formats': ('U10', 'f')})

# Esplora i dati
print("Prime 5 righe del file CSV:")
print(data[:5])
print("\nNumero di righe e colonne nel file CSV:", data.shape)

# Calcola la temperatura media, massima e minima
temperatura_media = np.mean(data['Temperatura'])
temperatura_massima = np.max(data['Temperatura'])
temperatura_minima = np.min(data['Temperatura'])
print("\nTemperatura media registrata:", temperatura_media)
print("Temperatura massima registrata:", temperatura_massima)
print("Temperatura minima registrata:", temperatura_minima)

# Calcola la deviazione standard della temperatura
deviazione_standard = np.std(data['Temperatura'])
print("\nDeviazione standard della temperatura:", deviazione_standard)

# Trova la data della temperatura massima e minima
data_temperatura_massima = data[np.argmax(data['Temperatura'])]['Data']
data_temperatura_minima = data[np.argmin(data['Temperatura'])]['Data']
print("\nData della temperatura massima registrata:", data_temperatura_massima)
print("Data della temperatura minima registrata:", data_temperatura_minima)

# Visualizza un grafico della temperatura nel tempo
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['Data'], data['Temperatura'], marker='o', linestyle='-', color='b')
plt.xlabel('Data')
plt.ylabel('Temperatura')
plt.title('Temperatura nel tempo')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
